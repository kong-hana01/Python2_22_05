### Class를 사용한 노드 만들기
기본적인 노드에서 각 개별 노드는 다음과 같은 정보를 가지고 있습니다.
- 변수: value, pointer
- 메소드: value 수정, value 검색, pointer 수정 등등

```python
class Node:
    def __init__(self, value = None, pointer = None):
        self.value = value
        self.pointer = pointer
        
    def getData(self):
        return self.value
    
    def setData(self, newData):
        self.value = newData
```

```python
n2 = Node()
```

```python
n1 = Node("a", None)
```

```python
n1.getData()
```

```python
n1.setData("b")
```

```python
n1.getData() 
```

```python
class Node:
    cnt = 0
    
    def __init__(self, value = None, pointer = None):
        self.value = value
        self.pointer = pointer
        Node.cnt += 1
    def getData(self):
        return self.value
    
    def setData(self, newData):
        self.value = newData
```

```python
n3 = Node()
```

```python
Node.cnt
```

```python
n4 = Node("dog", Node("cat", None))
```

```python
n4.value
```

```python
n4.pointer
```

```python
n4.pointer.value
```
