## requests 모듈

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.naver.com"
```

```python
# get 방식으로 해당 url에 request 하고 response 받아 리턴
response = requests.get(url)
response             # out 값 200이 나오면 정상적인 response code

# request 결과코드
# 200 성공
# 404 url 존재하지 않음
# 500 서버 내부 
# 400 Bad Request : request 문법상의 오류
# 401 권한에러 : 권한 관현 적절한 header 정보가 없는 경우 많이 발생
# 402 권한에러 : 자원에 대한 접근 권한 에러
# 403  권한에러 Forbidden : 파일권한, 읽기권한, SSL, IP, 등...  <--- 웹 크롤링 하다가 은근히 자주 마주치게 되는 에러
```

##  완성코드

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.naver.com"
response = requests.get(url)
dom = BeautifulSoup(response.text, "html.parser")
elements = dom.select("#yna_rolling .issue")
result = [
   element.text.strip()
    for element in elements
]
print(result)
```

###  결과 파일 저장하기 --> txt

```python
with open(filepath, "wt", encoding="utf-8") as f:
    for data in result:
        f.write(data + "\n")
```
